### Soluzione 1.5

```python
temperature = np.array([22, 24, 19, 21, 25, 28, 30])
giorni = ['Lun', 'Mar', 'Mer', 'Gio', 'Ven', 'Sab', 'Dom']

media = temperature.mean()
giorno_caldo_idx = np.argmax(temperature)
giorni_sopra_media = (temperature > media).sum()

print(f'Temperatura media: {media:.2f}°C')
print(f'Giorno più caldo: {giorni[giorno_caldo_idx]} con {temperature[giorno_caldo_idx]}°C')
print(f'Giorni sopra la media: {giorni_sopra_media}')

plt.bar(giorni, temperature, color='orange')
plt.axhline(y=media, color='red', linestyle='--', label=f'Media: {media:.1f}°C')
plt.title('Temperature settimanali')
plt.ylabel('Temperatura (°C)')
plt.legend()
plt.show()
```

### Soluzione SFIDA FINALE

```python
def f(x):
    return (x - 3)**2 + 2

def derivata(x):
    return 2 * (x - 3)

x = 10.0  # punto di partenza
learning_rate = 0.1
storia = [x]

for i in range(30):
    x = x - learning_rate * derivata(x)
    storia.append(x)

print(f'Valore finale di x: {x:.4f} (atteso: 3)')
print(f'Valore finale di f(x): {f(x):.4f} (atteso: 2)')

plt.plot(storia, marker='o')
plt.axhline(y=3, color='red', linestyle='--', label='minimo (x=3)')
plt.title('Discesa del gradiente: x si avvicina al minimo')
plt.xlabel('Iterazione')
plt.ylabel('Valore di x')
plt.legend()
plt.grid(True)
plt.show()
```
