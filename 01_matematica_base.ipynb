## 1️⃣ Variabili e funzioni

Una **variabile** è semplicemente un nome a cui associamo un valore. In Python si scrive così:

```python
# Una variabile: un "contenitore" che ha un nome (x) e un valore (5)
x = 5
print("Il valore di x è:", x)

# Posso cambiare il valore
x = 10
print("Adesso x vale:", x)
```

Una **funzione** è una "macchinetta" che prende un input e restituisce un output.

Esempio matematico: $f(x) = 2x + 3$ 
- se metti dentro `x = 1`, esce `5`
- se metti dentro `x = 4`, esce `11`

Vediamolo in Python:

```python
# Definiamo la funzione f(x) = 2x + 3
def f(x):
    return 2 * x + 3   # restituisce il risultato

# Proviamola con vari valori
print("f(1) =", f(1))
print("f(4) =", f(4))
print("f(0) =", f(0))

# 🔧 PROVA TU: cambia il numero dentro f() e vedi cosa succede
```

💡 **Concetto chiave**: nel Machine Learning, il modello è **una funzione**. 
Riceve in input le caratteristiche di qualcosa (es: età, peso, altezza) e restituisce una previsione (es: probabilità di una malattia).

Ora **disegniamo** la funzione per visualizzarla. Useremo `matplotlib`, la libreria standard per i grafici.

```python
# Importiamo le librerie
import numpy as np              # numpy serve per gestire numeri e array in modo veloce
                                #   (Un array è una struttura dati fondamentale nella programmazione
                                #   che agisce come un contenitore unico per memorizzare una sequenza
                                #   di elementi dello stesso tipo (es. solo numeri o solo testo)
                                #   in posizioni di memoria adiacenti. Si può immaginare come una griglia
                                #   o una fila di scatole numerate, dove ogni scatola contiene un dato.

import matplotlib.pyplot as plt # matplotlib serve per disegnare grafici

# Creiamo 100 valori di x tra -10 e 10
x = np.linspace(-10, 10, 100)   # linspace = "spazio lineare"
y = 2 * x + 3                    # calcoliamo y per ogni x

# Disegniamo il grafico
plt.figure(figsize=(8, 5))       # dimensione del grafico
plt.plot(x, y, color='blue', label='f(x) = 2x + 3')
plt.axhline(0, color='black', linewidth=0.5)  # asse orizzontale
plt.axvline(0, color='black', linewidth=0.5)  # asse verticale
plt.grid(True, alpha=0.3)
plt.title('Una funzione lineare')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()

# 🔧 PROVA TU: cambia 2*x + 3 in qualcos'altro, ad esempio x**2 (parabola)
```

## 2️⃣ Vettori e matrici

Un **vettore** è una lista ordinata di numeri. Esempio: i tuoi voti in 5 materie:
$$ v = [7, 8, 6, 9, 7] $$

![schema dei vettori](Media/01/vettore.png)

Una **matrice** è una tabella di numeri (vettori messi uno sotto l'altro). Esempio: i voti di 3 studenti in 5 materie.

![schema matrice](Media/01/matrice.png)


Nel Machine Learning, **i dati sono quasi sempre matrici**: ogni riga è un esempio, ogni colonna una caratteristica.



```python
import numpy as np

# Un vettore (1 dimensione)
voti = np.array([7, 8, 6, 9, 7])
print("Vettore dei voti:", voti)
print("Numero di elementi:", voti.shape)  # shape = forma

# Una matrice (2 dimensioni): 3 studenti × 5 materie
voti_classe = np.array([
    [7, 8, 6, 9, 7],   # studente 1
    [5, 6, 7, 8, 6],   # studente 2
    [9, 9, 8, 10, 9],  # studente 3
])
print("\nMatrice dei voti della classe:")
print(voti_classe)
print("Forma (righe, colonne):", voti_classe.shape)
```

### Operazioni utili sui vettori e matrici

Queste operazioni sono **il pane quotidiano del ML**: somma, media, prodotto.

```python
# Somma di tutti i voti di uno studente
print("Somma voti studente 1:", voti.sum())

# Media (importantissima nel ML!)
print("Media voti studente 1:", voti.mean())

# Massimo e minimo
print("Voto massimo:", voti.max())
print("Voto minimo:", voti.min())

# Sulla matrice possiamo calcolare la media per riga (axis=1) o per colonna (axis=0)
print("\nMedia per studente:", voti_classe.mean(axis=1))   # asse 1 = lungo le colonne, una riga alla volta
print("Media per materia:", voti_classe.mean(axis=0))      # asse 0 = lungo le righe, una colonna alla volta

# 🔧 PROVA TU: calcola la somma dei voti per ogni studente
```

### Il prodotto scalare (dot product)

È un'operazione fondamentale: moltiplica elemento per elemento e somma tutto.
$$ a \cdot b = a_1 b_1 + a_2 b_2 + ... + a_n b_n $$

**Perché è importante?** Quasi tutti i modelli di ML calcolano previsioni con un prodotto scalare!

Esempio concreto: vogliamo calcolare la **media pesata** di 3 esami con pesi diversi:

```python
# Voti di 3 esami
voti = np.array([7, 8, 9])

# Pesi: il primo esame conta 50%, il secondo 30%, il terzo 20%
pesi = np.array([0.5, 0.3, 0.2])

# Prodotto scalare = media pesata
media_pesata = np.dot(voti, pesi)   # equivale a 7*0.5 + 8*0.3 + 9*0.2
print("Media pesata:", media_pesata)

# 💡 Questa è ESATTAMENTE la formula della Regressione Lineare:
#    previsione = peso_1 * caratteristica_1 + peso_2 * caratteristica_2 + ...
# Il modello impara i "pesi" giusti per fare buone previsioni!
```

## 3️⃣ La derivata: come "misurare il cambiamento"

La **derivata** di una funzione in un punto ci dice **quanto velocemente** la funzione sta crescendo o diminuendo in quel punto.

🚗 Analogia: se in macchina il contachilometri è la posizione, il **tachimetro** è la derivata (la velocità).
![schema derivata](Media/01/derivata.png)

**Perché serve nel ML?** 
Il modello fa previsioni sbagliate → calcoliamo l'**errore** → la derivata ci dice **in quale direzione modificare i pesi** per ridurre l'errore. Questo si chiama **discesa del gradiente** ed è il cuore del ML moderno!

Vediamolo graficamente:

```python
import numpy as np
import matplotlib.pyplot as plt

# Funzione: f(x) = x^2 (una parabola)
def f(x):
    return x**2

# La sua derivata, calcolata a mano, è 2x
def derivata(x):
    return 2 * x

x = np.linspace(-5, 5, 100)
y = f(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'b-', label='f(x) = x²')

# Disegniamo la "pendenza" (derivata) in 3 punti
for punto in [-3, 0, 2]:
    pendenza = derivata(punto)
    # disegniamo una piccola retta tangente con quella pendenza
    xx = np.linspace(punto - 1, punto + 1, 10)
    yy = f(punto) + pendenza * (xx - punto)
    plt.plot(xx, yy, 'r-', linewidth=2)
    plt.plot(punto, f(punto), 'ro', markersize=8)
    plt.text(punto, f(punto) - 2, f'pendenza={pendenza}', fontsize=10, ha='center')

plt.title('La derivata = la pendenza della retta tangente')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# 💡 Nota:
# - in x = -3 la pendenza è negativa (-6) → la funzione sta scendendo
# - in x = 0 la pendenza è 0 → siamo nel "fondo" (minimo)
# - in x = 2 la pendenza è positiva (4) → la funzione sta salendo
# Per arrivare al minimo dovremmo andare nella direzione OPPOSTA alla pendenza!
# Questa è l'idea della discesa del gradiente.
```

### 🎯 Mini esempio di Discesa del Gradiente

![schema gradiente](Media/01/gradiente.png)


Cerchiamo il minimo della funzione $f(x) = x^2$ partendo da un punto a caso, usando la derivata:

```python
# Punto di partenza casuale
x = 4.0

# Quanto "grande" è ogni passo (learning rate, lo vedremo molte volte!)
passo = 0.1

print(f"Partenza: x = {x}, f(x) = {f(x)}")

# Facciamo 20 "passi" verso il minimo
for i in range(20):
    grad = derivata(x)         # quanto è ripida la funzione qui?
    x = x - passo * grad        # ci muoviamo nella direzione opposta alla pendenza
    print(f"Passo {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")

print("\n🎉 Siamo arrivati vicini a x=0, il minimo della parabola!")

# 🔧 PROVA TU: cambia 'passo' a 0.5 o a 1.5 e guarda cosa succede!
# (con passi troppo grandi, l'algoritmo "rimbalza" e non converge)
```

## 4️⃣ Statistica di base

Il ML è anche statistica. I tre concetti fondamentali da conoscere:

- **Media**: il valore "centrale" tipico dei dati
- **Varianza**: quanto i dati sono "sparpagliati" intorno alla media
- **Deviazione standard**: la radice quadrata della varianza (più intuitiva, stessa unità di misura dei dati)

```python
import numpy as np

# Altezze (in cm) di 10 persone
altezze = np.array([170, 165, 180, 175, 168, 172, 178, 169, 174, 171])

# Media: somma di tutto / numero di elementi
media = altezze.mean()
print(f"Altezza media: {media} cm")

# Varianza: media dei quadrati delle distanze dalla media
varianza = altezze.var()
print(f"Varianza: {varianza:.2f}")

# Deviazione standard: "quanto in media le persone si discostano dalla media"
dev_std = altezze.std()
print(f"Deviazione standard: {dev_std:.2f} cm")
print(f"\nIn media le altezze stanno tra {media - dev_std:.1f} e {media + dev_std:.1f} cm")
```

### Confronto: dati "compatti" vs dati "sparpagliati"

Due gruppi possono avere la **stessa media** ma essere molto diversi!

```python
# Gruppo A: tutti molto simili
gruppo_A = np.array([99, 100, 101, 100, 100])

# Gruppo B: molto diversi tra loro
gruppo_B = np.array([50, 150, 80, 120, 100])

print("Gruppo A → media:", gruppo_A.mean(), "| dev. std:", round(gruppo_A.std(), 2))
print("Gruppo B → media:", gruppo_B.mean(), "| dev. std:", round(gruppo_B.std(), 2))

# Visualizziamoli
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].bar(range(len(gruppo_A)), gruppo_A, color='green')
axes[0].set_title(f'Gruppo A — dati compatti (std={gruppo_A.std():.1f})')
axes[0].set_ylim(0, 200)
axes[1].bar(range(len(gruppo_B)), gruppo_B, color='red')
axes[1].set_title(f'Gruppo B — dati sparpagliati (std={gruppo_B.std():.1f})')
axes[1].set_ylim(0, 200)
plt.show()

# 💡 Nel ML è cruciale capire quanto i dati "variano":
# tanta varianza = problema più difficile da prevedere
```

## 5️⃣ Distribuzioni di probabilità

Una **distribuzione** descrive **come si distribuiscono i valori** di una variabile.

La più famosa è la **distribuzione normale** (a campana, o "Gaussiana"):
- molti valori vicino alla media
- pochi valori molto lontani dalla media

Esempi reali: altezza delle persone, errori di misura, voti agli esami…

![schema normale](Media/01/normale.png)


```python
import numpy as np
import matplotlib.pyplot as plt

# Generiamo 1000 numeri casuali da una distribuzione normale
# media = 170, deviazione standard = 10  (come le altezze degli adulti)
np.random.seed(42)  # per ottenere sempre gli stessi numeri (riproducibilità)
altezze_simulate = np.random.normal(loc=170, scale=10, size=1000)

# Disegnamo l'istogramma
plt.figure(figsize=(10, 5))
plt.hist(altezze_simulate, bins=30, color='skyblue', edgecolor='black')
plt.axvline(170, color='red', linestyle='--', label='Media (170)')
plt.title('Distribuzione Normale — Altezze simulate')
plt.xlabel('Altezza (cm)')
plt.ylabel('Quante persone')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Media reale: {altezze_simulate.mean():.2f}")
print(f"Dev. std reale: {altezze_simulate.std():.2f}")

# 🔧 PROVA TU: cambia 'scale' (la dev. std) e guarda come cambia la "larghezza" della campana
```
